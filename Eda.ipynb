```python
def validate_hotel_data(number_of_rooms):
    # Rule: A hotel must have at least 1 room to exist!
    if number_of_rooms <= 0:
        # Intentionally force a ValueError exception to stop the script
        raise ValueError(f"Data Input Error: Invalid room count ({number_of_rooms}). Must be 1 or more.")
        
    print("Data looks clean! Passing variables to the AI model...")
```

```python
try:
    # A corrupt row from our dataset
    validate_hotel_data(-3)
except ValueError as e:
    # Your try/except block catches the error you raised intentionally!
    print(f"Pipeline Guard Alert: {e}")
```

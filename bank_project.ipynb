```python
class InsufficientFundsError(Exception):  pass
class InvalidAmountError(Exception) : pass
```

```python
class BankAccount :
    def __init__(self,owner,__balance = 100):
        self.owner = owner
        self.__balance = __balance
    def get_balance(self) : 
        return self.__balance
    def deposit(self, amount):
        if amount <= 0:
            raise InvalidAmountError("Your amount should be > 0$")
            
        self.__balance += amount
        print("💰 Your operation is done successfully")

    def withdraw(self, amount):
        if amount <= 0:
            raise InvalidAmountError("Your amount should be > 0$")
            
        if self.__balance - amount < 0:
            raise InsufficientFundsError(
                f"Your balance (${self.__balance}) doesn't have enough money to withdraw this amount."
            )
            
        self.__balance -= amount
        print("💸 Your operation is done successfully")
```

```python
def run_app():
    print("====================================")
    print("      WELCOME TO PYTHON BANK        ")
    print("====================================")
    
    name = input("Enter your name to open an account: ").strip()
    if not name:
        name = "Valued Guest"
    
    user_account = BankAccount(owner=name) 
    print(f"\n🎉 Welcome {user_account.owner}! Account created with a $100.00 starting balance.")
    
    while True:
        print("\n------------------------------------")
        print("1. Deposit Money")
        print("2. Withdraw Money")
        print("3. Check Balance")
        print("4. Exit Application")
        print("------------------------------------")
        
        choice = input("Select an option (1-4): ").strip()
        if choice == "1" :
            try : 
                amt = float(input("Enter deposit amount"))
                user_account.deposit(amt)
            except ValueError:
                print("🚨 Input Error: Please type digits only.")
            except InvalidAmountError as e :
                print(f"🚨 User Error: {e}")
        elif choice == "2":
            try:
                amt = float(input("Enter Withdraw amount"))
                user_account.withdraw(amt)
            except ValueError:
                print("🚨 Input Error: Please type digits only.")
            except InvalidAmountError as e1 :
                print(f"🚨 User Error: {e1}")
            except InsufficientFundsError as e2:
                print(f"🚨 User Error: {e2}")
        elif choice == "3":
            print(f"Your balance = {user_account.get_balance()}")
        elif choice == "4":
            print(f"\nThank you for banking with us, {user_account.owner}. Goodbye!")
            break 
        else:
            print("\n❌ Invalid Choice. Please pick a number from 1 to 4.")

if __name__ == "__main__":
    run_app()
                
```

# Pandas


```python
data = {
    "Age (Series 1)": [22, 35, 58],
    "City (Series 2)": ["New York", "Paris", "Tokyo"]
}

# 2. Define your custom row labels (The Index)
row_labels = ["Alice", "Bob", "Charlie"]

# 3. Feed them into the DataFrame constructor
df = pd.DataFrame(data, index=row_labels)

# 4. Display the DataFrame
df
```

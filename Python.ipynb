# matplotlib

```python

x = np.arange(10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c ='r')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph 2D")
plt.savefig('img11.png')
```

```python
y = x*x
```

```python
plt.plot(x,y,'go--')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,y,'r')


plt.tight_layout()
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("sin wave form")
plt.plot(x,y)
```

```python
x = np.arange(10)
y = np.arange(11,21)
plt.bar(x,y)
plt.title("bar graph")
```

```python
a =np.array([20,20,30,50,17,60,60,60,60,8])
plt.hist(a,color = 'r')
plt.title("historgram")
```

```python
labels = ["python " , "java" , "C++", "C"]
sizes = [20,10,30,5]
colors = ["Red" , "Blue" , "Green" , "Yellow"]
plt.pie(sizes,explode = (0.1,0,0,0) ,labels = labels , colors = colors ,autopct ="%1.1f%%" ,shadow = True)
plt.axis('equal')
```

# seaborn


```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.corr(numeric_only = True)
```

```python
sns.heatmap(df.corr(numeric_only = True) , annot=True )
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

```python
# We remove kind='hex' so we can cleanly view the categorical color split
sns.jointplot(x='tip', y='total_bill', data=df, hue='smoker')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df , hue = 'smoker')
```

```python
sns.distplot(df['tip'] , kde = False , bins = 10)
```

```python
sns.countplot(x='day',data =df)
```

```python
sns.barplot(y = 'total_bill' , x = 'day' , data =df)
```

```python
sns.boxplot(x = 'sex' , y = 'total_bill' , data =df )
```

```python
sns.boxplot( x = 'total_bill', y = 'smoker' , data =df , hue ='day') 
```

```python
sns.violinplot( x = 'total_bill', y = 'smoker' , data =df ) 
```
